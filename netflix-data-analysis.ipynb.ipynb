```python
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
```

```python
df['year_added'] = df['date_added'].dt.year
```

```python
yearly_content = df['year_added'].value_counts().sort_index()
print(yearly_content)
```

```python
import matplotlib.pyplot as plt

yearly_content.plot(kind='line', marker='o')
plt.title("Year-wise Content Added on Netflix")
plt.xlabel("Year")
plt.ylabel("Number of Shows")
plt.show()
```

```python
**->->INSIGHT<-<-**
### 📌 Insight
Content addition increased rapidly after 2015, showing Netflix’s aggressive expansion strategy, with peak growth between 2018 and 2020.
```

```python
df['country'] = df['country'].str.split(',')
df = df.explode('country')
df['country'] = df['country'].str.strip()

india_df = df[df['country'] == 'India']
india_yearly = india_df.groupby('year_added').size()
print(india_yearly)
print(india_df.head())
```

```python
**->->INSIGHT<-<-**

Indian content on Netflix increased rapidly from 2016 to 2018, reaching a peak in 2018, after which it showed a slight decline, indicating an initial expansion followed by stabilization.
```

## 📈 Key Insights
