```python
df=df.drop_duplicates()
```

```python
df.duplicated().sum()
```

```python
import matplotlib.pyplot as plt

total_sales = df['sales'].sum()
total_profit = df['profit'].sum()

print("Total Sales:", total_sales)
print("Total Profit:", total_profit)

labels = ['Sales', 'Profit']
values = [total_sales, total_profit]

plt.bar(labels, values)
plt.title("Total Sales vs Profit")
plt.xlabel("Category")
plt.ylabel("Amount")
plt.show()
```

## INSIGHT##
The company has generated high overall sales, but the profit is relatively low in comparison.

```python
df.groupby('category')['profit'].sum().plot(kind='bar')

import matplotlib.pyplot as plt
plt.title("Profit by Category")
plt.xlabel("Category")
plt.ylabel("Profit")
plt.show()
```

##INSIGHT##
Technology category generates the highest profit among all categories.

```python
region_profit = df.groupby('region')['profit'].sum()


region_profit.plot(kind='pie', autopct='%1.1f%%', explode=[0.1,0,0,0])

plt.title("Profit Distribution by Region")
plt.ylabel('')

plt.show()
```

##INSIGHT##
The West region generates the highest profit among all regions.
